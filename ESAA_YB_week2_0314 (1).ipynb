# [개념정리]
파이썬 기반 시각화의 대표적인 라이브러리 : 맷플롯립 & 시본

---

**맷플롯립의 문제점**
1) 차트의 축 이름, 차트 타이틀, 범례 등의 부가적인 속성까지 코딩해주어야 함
2) 기본 설정 환경에서는 현대적인 감각이 떨어지는 시각화 플롯 등의 문제점

**시본**
* 맷플롯립보다 쉬운 구현, 수려한 시각화, 그리고 판다스와의 연동
* 맷플롯립을 기반으로 함 -> 기본적인 수준의 이해가 반드시 필요
* 맷플롯립보다 적은 양의 코딩으로 수려한 시각화 플롯 제공

**맷플롯립**
* pyplot을 제공하며 이를 통해 시각화 구현 가능
* matplotlib.pyplot 모듈 -> plt란 이름으로 import
* plot() : 선 그래프를 그릴 수 있음 -> x와 y값이 매핑되는 선 그래프를 그릴 수 있음
* title() : 입력값으로 타이틀 문자열을 받아서 차트 타이틀 설정
* show() : 축을 포함한 선 그래프를 화면에 나타냄
* X축과 Y축의 명칭을 설정하거나 범례를 설정하는 작업을 모두 별도의 함수들을 사용해서 구현해야함

**figure: 캔버스 역할** 
* 실제적으로 그림을 그리는 역할을 수행 X
* 주로 figure의 크기를 조절하는데 사용
* figsize=(가로크기, 세로크기)입력
* facecolor -> 그림판의 배경 색깔 설정


**axes : 실제 그림을 그리는 메서드들 가짐** 
* x축, y축, 타이틀, 범례 등의 설정하는데 용이

**plot.subplots()**
* pyplot에서 figure와 axes 객체를 함께 가져올 수 있음
* 주로 여러개의 subplot을 생성하는데 활용함.->이때 개별 subplot은 하나의 Axes를 가지게 됨.
* subplots()의 인자 nrows, ncols, figsize
* nrows : 행의 개수 & ncols : 열의 개수
* figsize : 모든 subplot들을 포함한 전체 Figure의 크기를 설정

**pyplot의 plot()함수**
1) 선 그래프를 그릴때 활용 -> x좌표, y좌표 값으로 파이썬 리스트, 넘파이 ndarray, 판다스의 DataFrame/Series 모두 적용 가능
   단) x좌표 값과 y 좌표 값의 개수는 모두 같은 크기를 가져야함.
2) color 인자를 이용해 선의 색깔을 변경할 수 있음

**축 & 범례 설정**
* xlabels(), ylabels() : 축의 명칭 설정
* xticks() : x축의 눈금값을 회전시키기 위해
* yticks() : y축의 눈금값을 회전시키기 위해
* if 축의 값이 문자열 -> 모든 값을 나타냄
* if 축의 값이 숫자값 -> 최솟값과 최댓값에 기반해 축의 크기에 따라 눈금 간격을 만들어줌
* 범례설정
  1) plt.plt() 함수 내에 lable 인자값을 설정
  2) plt.legend()를 호출해 범례를 axes 상에 나타냄.
* 직접 axes객체 메서드를 활용해 시각화
  1) plt.axes()를 이용해 axes 객체를 추출
  2) plt.plot() -> axes객체의 plot(), plt.bar() -> axes 객체의 bar(), plt.xlabl() -> axes 객체의 set_xlabel(),plt.ylabl() -> axes 객체의 set_ylabel() 변경
  3) plt.title() ->axes 객체의 set_title()로 변경
  4) plt.legend -> legend로 적용

**개별 그래프들을 subplot별로 시각화하기**
* 변환된 axes 객체들을 가지는 ax 변수를 튜플이 아니라 ndarray 형태를 가짐
* 유의 사항: 개별 axes 객체를 직접 이용해서 시각화를 구현해야한다. -> 타이틀, 축의 명칭, 눈금값, 범례 설정할때에도 개별 axes에 직접 이들을 적용해야함.
* if plt.subplots()의 인자인 nrows=1 or ncols=1 -> 1차원 배열 인덱싱 방식으로 axes를 접근할 수 있음
* if nrows>=2 or ncols>=2 -> 2차원 배열 인덱싱 방식으로 접근

# [코드 필사]

```python
#pyplot 모듈을 (관용적으로) plot라는 이름으로 import.
import matplotlib.pyplot as plt

#x축 입력각 [1,2,3], y축 입력값 [2,4,6]으로 선 그래프 생성.
plt.plot([1,2,3],[4,5,6])

#선 그래프 타이틀을 Hello plot으로 설정
plt.title("Hello plot")

# 선 그래프를 출력
plt.show()
```

```python
#plt.figure()는 주로 figure의 크기를 조절하는데 사용됨.
# 아래는 figure 크기가 가로가 10, 세로 4인 Figure 객체를 설정하고 반환함.
plt.figure(figsize=(10,4))

plt.plot([1,2,3],[4,5,6])
plt.title("Hello plot")
plt.show
```

```python
plt.figure(figsize=(4,4), facecolor='yellow')
plt.plot([1,2,3],[2,4,6])
plt.title("Hello plot")
plt.show()
```

```python
ax=plt.axes()
print(type(ax))
```

```python
#pyplot에서 설정된 figure와 axes 객체를 함께 가져오기
fig, ax=plt.subplots()
print('fig type:',type(fig), '\nax type:', type(ax))
```

```python
fig, ax= plt.subplots(figsize=(4,4))

#Axes 객체의 plot()을 이용하여 선 그래프를 그림.
ax.plot([1,2,3],[2,4,6])

#Axes 객체의 set_title()을 이용하여 선 그래프 타이틀을 Hello plot으로 설정.
ax.set_title('Hello plot')
plt.show()
```

```python
# 아래는 두개의 subplot, 즉 2개의 Axes를 생성하되 열 방향으로 두개를 배치함.
# 또한 전체 크기가 가로6, 세로 3인 figure 생성.
fig,(ax1, ax2)= plt.subplots(nrows=1, ncols=2, figsize=(6,3))
```

```python
x_value=np.arange(0,100)
y_value=2*x_value

plt.plot(x_value,y_value, color='green')
plt.xlabel('x axis')
plt.ylabel('y axis')

#x축의 눈금값을 np.arange(0,100,5)로 하여 0부터 99까지 5스텝으로 20개를 표시하고
#90도 회전
plt.xticks(ticks=np.arange(0,100,5),rotation=90)

plt.title('Hello plot')
plt.show()
```

```python
#x좌표 값은 0~99까지 100개, 첫번째 Y좌표값은 x좌표값의 두배, 두번째 좌표값은 x좌표값의 4배
x_value_01=np.arange(0,99)
y_value_01=2*x_value_01
y_value_02=4*x_value_01

#x좌표값과 첫번째 y좌표값으로 녹색의 선그래프를 그림. label을 'green line'으로 설정
plt.plot(x_value_01, y_value_01, color='green', label='green line')
#x좌표값과 두번째 y좌표값으로 빨간색의 선그래프를 그림. label을 'red line'으로 설정
plt.plot(x_value_01, y_value_02, color='red', label='red line')

plt.xlabel('x_axis')
plt.ylabel('y_axis')
# 개별 plot들의 label로 설정된 문자열을 범례로 표시
plt.legend()

plt.title('Hello plot')
plt.show()
```

```python
x_value_01=np.arange(0,10)
y_value_01=2*x_value_01

#마커를 포함한 빨간색 대신 선 그래프를 그리고, label을 'red line으로 설정
plt.plot(x_value_01, y_value_01, color='red', marker='o', linestyle='dashed',label='red line')
#x값에 따른 y값을 나타내는 초록색 막대 그래프를 그리고 label은 'bar plot'으로 설정
plt.bar(x_value_01, y_value_01, color='green', label='bar plot')

plt.xlabel('x_axis')
plt.ylabel('y_axis')
# 개별 plot들의 label로 설정된 문자열을 범례로 표시
plt.legend()

plt.title('Hello plot')
plt.show()
```

```python
figure=plt.figure()
#Axes 객체를 추출하여 ax 변수에 할당
ax=plt.axes()

#plt.plot()은 ax.plot()으로 plt.bar()은 ax.bar()로 변경
ax.plot(x_value_01, y_value_01, color='red', marker='o', linestyle='dashed',label='red line')
ax.bar(x_value_01, y_value_01, color='green', label='bar plot')

#plt.xlabel()은 ax.set_xlabel()로, plt.ylable()은 ax.set_ylabel로 변경
ax.set_xlabel('x_axis')
ax.set_ylabel('y_axis')

#ax.set_legend가 아니라 ax.legend()임.
ax.legend

#plt.title()을 ax.set_title()로 변경
ax.set_title('Hello plot')

plt.show()
```

```python
import numpy as np

x_value_01=np.arange(1,10)
x_value_02=np.arange(1,20)
y_value_01=2*x_value_01
y_value_02=2*x_value_02

#figsize는 (6,3)인 Figure와 2개의 Axes를 가지는 subplots을 반환
fig, ax=plt.subplots(nrows=1,ncols=2,figsize=(6,3))

# 개별 Axes 객체들을 각각 이용하여 선 그래프와 막대그래프를 그림.
ax[0].plot(x_value_01, y_value_01, color='red', marker='o', linestyle='dashed',label='red line')
ax[1].bar(x_value_02, y_value_02, color='green', label='bar plot')

# 개별 Axes 객체에 각각 x축명을 설정.
ax[0].set_xlabel('ax[0] x_axis')
ax[1].set_xlabel('ax[1] y_axis')

# 개별 Axes 객체 각각에 범례 설정
ax[0].legend()
ax[1].legend()

# 개별 Axes 객체 각각에 타이틀을 설정
ax[0].set_title('Hello line')
ax[1].set_title('Hello bar')

plt.show()
```

```python
import numpy as np

x_value_01=np.arange(1,10)
x_value_02=np.arange(1,20)
y_value_01=2*x_value_01
y_value_02=2*x_value_02

#figsize는 (6,6)인 Figure와 2*2개의 Axes를 가지는 subplots을 반환
fig, ax=plt.subplots(nrows=2,ncols=2,figsize=(6,6))

# 개별 Axes 객체들을 각각 이용하여 선 그래프와 막대그래프를 그림.
ax[0][0].plot(x_value_01, y_value_01, color='red', marker='o', linestyle='dashed',label='red line')
ax[0][1].bar(x_value_02, y_value_02, color='green', label='bar bar')
ax[1][0].plot(x_value_01, y_value_01, color='green', marker='o', linestyle='dashed',label='green line')
ax[1][1].bar(x_value_02, y_value_02, color='red', label='red bar')

# 개별 Axes 객체에 각각 x축명을 설정.
ax[0][0].set_xlabel('ax[0][0] x_axis')
ax[0][1].set_xlabel('ax[0][1] x_axis')
ax[1][0].set_xlabel('ax[1][0] x_axis')
ax[1][1].set_xlabel('ax[1][1] x_axis')

# 개별 Axes 객체 각각에 범례 설정
ax[0][0].legend()
ax[0][1].legend()
ax[1][0].legend()
ax[1][1].legend()

plt.show()
```
