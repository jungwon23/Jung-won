# [개념정리]

**판다스**
*   행과 열로 이루어진 2차원 데이터를 효율적으로 가공/처리할 수 있는 다양하고 훌륭한 기능을 제공
*   내부 데이터뿐만 아니라 CSV등의 파일을 쉽게 DataFrame으로 변경해 데이터의 가공/분석을 편리하게 수행
*   핵심 객체: DataFrame
*   Index : 개별 데이터를 고유하게 식별하는 Key값
*   Series(칼럼이 하나뿐인 데이터 구조체)와 DataFrame(칼럼이 여러개인 데이터 구조체) : Index를 key값으로 가짐

---

**판다스 시작-파일을 DataFrame으로 로딩, 기본API**

*   1) 새로 주피터 노트북으로 생성하고 판다스 모듈 임포트하는 것
*   2) 데이터 파일을 판다스의 DataFrame으로 로딩할 것
*  shape: DataFrame의 행과 열을 튜플 형태로 반환
*  info() : 총데이터건수, 데이터 타입, Null건수
*  describe(): 칼럼별 숫자형 데이터값의 n-percentile분포도, 평균값, 최댓값,최솟값

 : 오직 숫자형(int, float 등) 칼럼의 분포도만 조사
*  value_counts() 
: 해당 칼럼값의 유형과 건수를 확인, 많은 건수 순서로 정렬되어 값을 반환

: 반환하는 데이터 타입-> series객체

:dropna의 기본값은 True -> Null 값을 무시하고 개별 데이터 값의 건수 계산

**DataFrame과 리스트, 딕셔너리, 넘파이 ndarray 상호 변환**

*   DataFrame:칼럼명 있음 (리스트와 넘파이 ndarray와 차이점)
*   1차원 형태의 리스트,넘파이 ndarray->DataFrame : 칼럼은 1개만 필요함
*   딕셔너리 -> DataFrame : key는 문자열 컬럼명, value는 리스트형 컬럼 데이터로
*   df_dict.values : DataFrame -> ndarray로 변환시킴
*   df_dict.values.toilst() : dataframe->리스트로 변환시킴
*   df_dfict.to_dict() : dataframe -> 딕셔너리로 변환시킴

**DataFrame의 칼럼 데이터 세트 생성과 수정**
*   []연산자 이용
*   DataFrame의 새로운 칼럼을 추과하고 일괄적으로 0값을 할당

**DataFrame 데이터 삭제**
*   drop() 이용
*   DataFrame.drop(lables=None, axis=0, index=None, columns=None, level=None, inplace=False, errors='raise') -> 원형
*   +) axis 0: 로우 방향 축, axis 1: 칼럼방향 축
*   Inplace=False : 자기 자신의 DataFrame의 데이터는 삭제하지 않으며, 삭제된 결과 DataFrame을 반환
*   Inplace=True : 자신의 DataFrame의 데이터 삭제

**Index 객체**
*   DataFrame, Series에서 Index 객체만 추출하려면 DataFrame.index 또는 Series.index 속성을 통해 가능함
*   Index 객체 : 식별성 데이터를 1차원 array로 가지고 있음
*   한번 만들어진 DataFrame 및 Series의 Index 객체는 함부로 변경할 수 없음
*   연산에서 제외됨
*   reset_index() : 새롭게 인덱스를 연속 숫자형으로 할당해줌.
  -> 인덱스가 연속된 숫자형 데이터가 아닌 경우 -> 이를 연속 int 숫자형 데이터로 만들때
*   drop=True로 설정 : 기존 인덱스는 새로운 칼럼으로 추가되지 않고 삭제됨

**데이터 셀력션 및 필터링**
*  DataFrame['컬럼명'] <- 숫자를 대입하면 오류가 발생함
*  DataFrame['불링 인덱싱'] : 불링 인덱싱 표현도 가능함
*  DataFrame[0:2] : 슬라이싱 연산 사용X
*  iloc[] : 위치 기반 인덱싱 방식
 -> iloc[행위치 정수값, 열 위치 정수값]
 -> iloc[]: 불링 인덱싱을 제공 X
*  loc[] : 명칭 기반 인덱싱 방식
 -> loc[인덱스값, 칼럼명]과 같은 형식으로 데이터 추출
 -> loc[]에 슬라이싱 기호 적용 : 종료값 -1 이 아니라 종료값까지 포함하는 것을 의미함
 -> 불링 인덱싱 가능
*  DataFrame[['Name','Age']] -> 칼럼이 2개 이상인 경우
*  and : &, or: |, Not: ~

**정렬, Aggregation 함수, GroupBy 적용**
* sort_values() : 주요 파라미터는 by,ascending, inplace
* 1) by : 특정 칼럼을 입력하면 해당 칼람으로 정렬 수행
  2) ascending=True : 오름차순, ascending=False : 내림차순
  3) inplace=False : 호출한 DataFrame은 그대로 유지, 정렬된 DataFrame을 결과로 반환
* aggregation 함수 : min(), max(), sum(), count()
  -> count() : Null 값을 반영하지 않은 결과를 반환
* groupby() : 입력 파라미터 by에 칼럼을 임력하면 대상 칼럼으로 groupby 됨
  - DataFrame에 groupby()를 호출하여 반환된 결과에 aggregation 함수를 호출하면 groupby() 대상 칼럼을 제외한 모든 칼럼에 해당 aggregation 함수 적용
* agg() : 입력값으로 딕셔너리 형태로 aggregation이 적용될 칼럼과 aggregation 함수를 입력

**결손 데이터 처리하기**
* isna() : NaN 여부를 확인하는 API -> 모든 칼럼의 값이 Nan인지 아닌지 True나 False로 알려줌.
* fillna(): Nan값을 다른 값으로 대체하는 API
  -> inplace=True : 실제 데이터 세트 값이 변경된다!

**apply lambda 식으로 데이터 가공**
* apply lambda : DataFrame이나 Series의 레코드 별로 데이터를 가공하는 기능
* 함수명과 입력인자를 먼저 선언 -> 함수내에서 입력인자를 가공 -> 결과값을 return과 같은 문법으로 반환
  -> map() 함수를 결합해서 상용 if 여러 개의 값을 입력인자로 사용하는 경우
* if절의 경우 if 식보다 반환값을 먼저 기술
* else if는 지원하지 않음

```python
# 나이에 따라 세분화된 분류를 수행하는 함수 생성. 
def get_category(age):
    cat=''
    if age <=5 : cat='Baby'
    elif age <=12 : cat='Child'
    elif age <=18 : cat='Teenager'
    elif age <=25 : cat='Student'
    elif age <=35 : cat='Young Adult'
    elif age <=60 : cat='Adult'
    else : cat='Elderly'

    return cat

# lambda 식 위에서 생성한 get_category() 함수를 반환값으로 지정.
# get_category(x)는 입력값으로 'Age' 칼럼 값을 받아서 해당하는 cat 반환
titanic_df['Age_cat']=titanic_df['Age'].apply(lambda x: get_category(x))
titanic_df[['Age','Age_cat']].head()
```
