```python
# 데이터 로드
from sklearn.datasets import fetch_openml
mnist=fetch_openml('mnist_784', version=1)
```

```python
# train, test split
X_train = mnist['data'][:60000]
y_train = mnist['target'][:60000]
X_test = mnist['data'][60000:]
y_test = mnist['target'][60000:]
```

* 랜덤포레스트에 적용

```python
# 랜덤포레스트에 적용 (n_estimators=10, random_state=42)
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=10, random_state=42)
rf.fit(X_train, y_train)
```

```python
# 정확도 결과
from sklearn.metrics import accuracy_score
y_pred=rf.predict(X_test)
accuracy=accuracy_score(y_test, y_pred)
print('RF의 정확도는 ', accuracy)
```

# 1. PCA

* PCA로 변환한 데이터 세트에 랜덤 포레스트 적용

```python
# PCA 라이브러리 불러오기
from sklearn.decomposition import PCA
```

```python
# PCA 적용 95% : n_components=0.95
pca=PCA(n_components=0.95)
```

```python
# PCA 이후 랜덤포레스트 훈련
pca.fit(X_train)
X_train_pca=pca.transform(X_train)

rf_2=RandomForestClassifier(n_estimators=10, random_state=42)
rf_2.fit(X_train_pca, y_train)
```

```python
# 정확도 결과 (X_test도 PCA 적용해야합니다.)
X_test_pca=pca.transform(X_test)
y_pred=rf_2.predict(X_test_pca)
accuracy=accuracy_score(y_test, y_pred)
print('accuracy는', accuracy)
```

# 2. LDA

- LDA로 변환한 데이터 세트에 랜덤 포레스트 적용

```python
# LDA 라이브러리 불러오기
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
# LDA 적용 : n_components=5
lda=LinearDiscriminantAnalysis(n_components=5)
```

```python
# LDA 이후 랜덤포레스트 훈련
lda.fit(X_train,y_train)
X_train_lda=lda.transform(X_train)

rf_2=RandomForestClassifier(n_estimators=10, random_state=42)
rf_2.fit(X_train_lda, y_train)
```

```python
# 정확도 결과 (X_test도 LDA 적용해야합니다.)
X_test_lda=lda.transform(X_test)
y_pred=rf_2.predict(X_test_lda)
accuracy=accuracy_score(y_test, y_pred)
print('accuracy는', accuracy)
```
