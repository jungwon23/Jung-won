# 1. Lightgbm

* 데이터 '와인품질분류.zip'입니다.

1. 필요한 라이브러리를 가져옵니다.

```python
import pandas as pd
import lightgbm as lgbm
```

2. 데이터를 불러옵니다.

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
smpl_sub = pd.read_csv('sample_submission.csv')
```

```python
train.head()
```

```python
test.head()
```

3. 변수를 변환합니다.

```python
# type에는 white와 red 두 종류가 있습니다.
# 각각 0,1로 변환합니다.
train['type'] = train['type'].map({'red': 1, 'white': 0})
test['type'] = test['type'].map({'red': 1, 'white': 0})
```

```python
train.head()
```

```python
test.head()
```

```python
# 모델에 입력하기 전 데이터를 정형합니다.
# train_x : train의 'index', 'quality' 피처를 제거하고 저장합니다.
# train_y : train의 'quality'를 저장합니다.
# test_x : test의 'index'를 제거하고 저장합니다.
train_x=train.drop(['index','quality'], axis=1)
train_y=train['quality']
test_x=test.drop(['index'], axis=1)
```

```python
train_x.shape, train_y.shape, test_x.shape
```

```python
train_x.head()
```

```python
train_y.head()
```

```python
test_x.head()
```

4. LightGBM 모델을 생성 및 훈련합니다.

```python
from lightgbm import LGBMClassifier
lgbm_wrapper = LGBMClassifier()
lgbm_wrapper.fit(train_x, train_y)
```

5. 생성한 모델로 예측 데이터를 생성합니다.

```python
lgbm_pred=lgbm_wrapper.predict(test_x)
```

6. 제출파일을 생성합니다.

```python
smpl_sub.head()
```

```python
#서브미션 파일의 quality 변수에 예측값을 대입합니다.
smpl_sub['quality']=lgbm_pred
```

```python
smpl_sub.head()
```

```python
# csv 파일로 저장합니다.
smpl_sub.to_csv('submission.csv', index=False)
```

# 2. XGBoost

```python
# 데이터 다운로드 링크로 데이터를 코랩에 불러옵니다.

!wget 'https://bit.ly/3i4n1QB'

import zipfile
with zipfile.ZipFile('3i4n1QB', 'r') as existing_zip:
    existing_zip.extractall('data')
```

```python
# 라이브러리 및 데이터 불러오기
import pandas as pd
from xgboost import XGBClassifier

train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')

train.drop('index',axis = 1 ,inplace =True)
test.drop('index',axis = 1 ,inplace =True)
```

```python
# 데이터 확인
train.info()
```

```python
test.info()
```

```python
# 원핫 인코딩 (pd.get_dummies())
train=pd.get_dummies(data=train, columns=['type'], prefix=['type'],dtype=int)
test=pd.get_dummies(data=test, columns=['type'], prefix=['type'],dtype=int)
```

```python
# 모델 정의
from xgboost import XGBClassifier
xgb=XGBClassifier()
```

```python
# 모델 학습
# X 는 train에서 quality 를 제외한 모든 변수
# y 는 train의 qulity 변수
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(train['quality'])
X=train.drop('quality',axis=1)


# fit 메소드를 이용해 모델 학습
xgb.fit(X,y)
```

```python
# predict 메소드와 원핫 인코딩한 test 데이터를 이용해 품질 예측
xgb_pred=xgb.predict(test)
```

```python
# sample_submission.csv 파일을 불러와 예측된 값으로 채워 주기
submission= pd.read_csv('sample_submission.csv')
submission['quality'] = xgb_pred
```

```python
# 정답파일 내보내기
submission.to_csv('xgb_pred.csv',index = False)
```

# 3. 보팅

```python
# 데이터 다운로드 링크로 데이터를 코랩에 불러옵니다.

!wget 'https://bit.ly/3i4n1QB'

import zipfile
with zipfile.ZipFile('3i4n1QB', 'r') as existing_zip:
    existing_zip.extractall('data')
```

```python
# 라이브러리 및 데이터 불러오기
import pandas as pd
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier

from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier

train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')

train.drop('index',axis = 1 ,inplace =True)
test.drop('index',axis = 1 ,inplace =True)
```

```python
# 데이터 확인
train.info()
```

```python
# 원핫 인코딩 (pd.get_dummies())
train=pd.get_dummies(data=train, columns=['type'], prefix=['type'],dtype=int)
test=pd.get_dummies(data=test, columns=['type'], prefix=['type'],dtype=int)
```

```python
# 모델 정의 (LightGBM, XGBoost, Random Forest)
lgbm=LGBMClassifier()
xgb=XGBClassifier()
rf=RandomForestClassifier()
# VotingClassifier 정의
voting=VotingClassifier(estimators=[('lgbm', lgbm), ('xgb', xgb), ('rf', rf)])
```

```python
# 모델 학습

# X 는 train에서 quality 를 제외한 모든 변수
# y 는 train의 qulity 변수
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(train['quality'])
X=train.drop('quality',axis=1)


# fit 메소드를 이용해 모델 학습
voting.fit(X,y)
```

```python
# predict 메소드와 원핫 인코딩한 테스트 데이터를 이용해 품질 예측
test_pred=voting.predict(test)
```
