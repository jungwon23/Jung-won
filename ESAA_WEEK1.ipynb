###**넘파이**

-파이썬에서 선행대수 기반의 프로그램을 쉽게 만들 수 있게 지원

-배열 기반의 연산 + 다양한 데이터 핸들링 기능 제공

**ndarray**

: 넘파이의 기반 데이터 타입 -> 다차원 배열 생성 & 연산 수행

: ndarray 내에는 같은 데이터 타입만 가능하다 vs. 리스트는 서로 다른 데이터 타입을 가질 수 있음

:함수 인자로 dtype을 정해주지않으면 default는 float64임

-**array() 함수**: 다양한 인자를 입력받아 -> ndarray로 변환하는 기능

-s**hape 변수**: ndaray의 크기(행과 열의 수를 튜플형태로 가짐)

-**ndarray.ndim **: 각 array의 차원을 알 수 있음

-**dtype**을 이용해서 ndarray내의 데이터 타입을 알 수 있음

if) 서로 다른 데이터 타입이 섞여있는 경우 -> 데이터 타입이 더 큰 데이터 타입으로 변환됨.

-**astype()** : ndarray내 데이터 값의 타입 변경 -> 데이터 타입을 float에서 int로 바꿀경우 메모리를 절약할 수 있음.

-**arange()**: range()와 유사한 기능을 함, 0부터 함수 인자 값 -1까지의 값을 순차적으로 ndarray의 데이터값으로 변환해줌.

-**zeros()**: 모든 값을 0으로 채운 해당 shape를 가진 ndarray를 반환해줌

-**ones()**: 모든 값을 1으로 채운 해당 shape를 가진 ndarray를 반환해줌

-**resahpe()**
 : ndarray를 특정 차원 및 크기로 변환해줌, 저장된 사이즈로 변경이 불가능하면 오류가 발생

 :-1인자는 resahpe(-1,1)과 같은 상태로 자주 사용됨

 -**tolist()** : ndarray를 리스트 자료형으로 바꿀 수 있음.

 **인데싱(indexing)**

-단일 데이터값 추출을 제외하고 슬라이싱, 펜시인덱싱, 불린 인덱싱으로 추출된 데이터 세트는 모두 ndarray 타입임

**특정한 데이터만 추출**

 -단일값추출

 : [ndarray 객체에 해당하는 위치의 인덱스 값] 넣기

 if) 마이너스 기호를 사용하면 맨 뒤에서부터 데이터 추출

 : 2차원 데이터 접근의 차이는 로우와 컬럼 위치의 인덱스를 통해 접근

 : 로우와 컬럼은 넘파이 ndarray에서 사용하지X -> axis0(로우방향축)와 axis 1(컬럼방향축)

**슬라이싱**

:연속한 데이터를 슬라이싱해서 추출함

: ':'사이에 시작인덱스와 종료 인덱스를 표시해서 사용

: 2차원 ndarray에서 슬라이싱 할때 로우와 컬럼의 인덱스를 지칭하는 부분만 1차원과 다름

if) 2차원 ndarray에서 뒤에 오는 인덱스를 없애면 -> 1차원 ndarray로 반환
if) 3차원이면 -> 2차원으로 반환

**펜싱인덱싱**
: 리스트나 ndarray로 인덱스 집합을 정하면 해당 위치의 인덱스에 해당하는 ndarray를 반환하는 인덱싱 방법

**불린 인덱싱**
: 조건필터링과 검색을 동시에 할 수 있는 인덱싱 방식

: ndarray 객체를 인덱싱을 지정하는 []입력 -> False 값 무시 & True 값이 있는 위치 인덱스 값으로 자동 변환해 데이터 반환


---



**행렬**

1) **np.sort() **: 원행렬은 그대로 유지한채 원행렬의 정렬된 행렬 반환

2) **ndarray.sort()** : 원행렬 자체를 정렬한 형태로 변환하며 반환값은 None

+ 내림차순 : np.sort()[::-1]

3) **argsort()** : 원본 행렬이 정렬되었을때 기존 원본 행렬의 원소에 대한 인덱스를 필요로 할때 + 정렬행렬의 원본 행렬 index를 ndarray형으로 반환

**선형대수 연산**

-**np.dot()** : 두 행렬을 내적할때 이용함

-**transpose()** : 원행렬에서 행과 열 위치를 교환하는 원소로 구성된 전치행렬을 만들때 이용함

```python
org_array=np.array([3,1,9,5])
print('원본행렬:', org_array)
#np.sort()로 정렬
sort_array1=np.sort(org_array)
print('np.sort() 호출 후 반환된 정렬 행렬:', sort_array1)
print('np.sort() 호출 후 반환된 원본 행렬:', org_array)

#ndarray.sort()로 정렬
sort_array2=org_array.sort()
print('org_array.sort() 호출 후 반환된 정렬 행렬:', sort_array2)
print('org_array.sort() 호출 후 반환된 원본 행렬:', org_array)
```

```python
sort_array1_desc=np.sort(org_array)[::-1]
print('내림차순으로 정렬:', sort_array1_desc)
```

```python
array2d=np.array([[8,12],[7,1]])

sort_array2d_axis0=np.sort(array2d, axis=0)
print('로우 방향으로 정렬:\n', sort_array2d_axis0)


sort_array2d_axis1=np.sort(array2d, axis=1)
print('칼럼 방향으로 정렬:\n', sort_array2d_axis1)
```

```python
org_array=np.array([3,1,9,5])
sort_indices=np.argsort(org_array)
print(type(sort_indices))
print('행렬 정렬 시 원본 행렬의 인덱스:', sort_indices)
```

```python
org_array=np.array([3,1,9,5])
sort_indices_desc=np.argsort(org_array)[::-1]
print('행렬 내림차순 정렬 시 원본 행렬의 인덱스:', sort_indices_desc)
```

```python
import numpy as np

name_array=np.array(['John','Mike','Sarah','Kate','Samuel'])
score_array=np.array([78,95,84,98,88])

sort_indices_asc=np.argsort(score_array)
print('성적 오름차순 정렬 시 score_array의 인덱스:', sort_indices_asc)
print('성적 오름차순으로 name_array의 이름 출력:', name_array[sort_indices_asc])
```

```python
A=np.array([[1,2,3],[4,5,6]])
B=np.array([[7,8],[9,10],[11,12]])
dot_product=np.dot(A,B)
print('행렬 내적 결과:\n', dot_product)
```

```python
A=np.array([[1,2],[3,4]])
transpose_mat=np.transpose(A)
print('A의 전치행렬:\n', transpose_mat)
```
